```python
from diffusers import StableDiffusionDepth2ImgPipeline
from PIL import Image
import torch

# Step 1: Load the model
model_id = "stabilityai/stable-diffusion-2-depth"
pipe = StableDiffusionDepth2ImgPipeline.from_pretrained(model_id).to("cuda")

# Step 2: Load the depth map image
# depth_image_path = "depth_map.png"  # Update with your file path
depth_image_path = "Cat_org.png"  # Replace with the actual file name of your initial image
depth_image = Image.open(depth_image_path).convert("RGB")

# Step 3: Define prompts
prompt = "A serene forest landscape with a river and sunlight filtering through the trees"
negative_prompt = "No buildings, people, or artificial structures"

# Step 4: Generate the image
generated_image = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    image=depth_image,
    guidance_scale=7.5
).images[0]

# Step 5: Save the generated image
output_path = "generated_image.png"
generated_image.save(output_path)
print(f"Image saved at {output_path}")
```
