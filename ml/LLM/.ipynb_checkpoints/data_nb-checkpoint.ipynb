```python
# Load the dataset
lyrics = pd.read_csv('app/lyrics-data.csv')

# Print the shape of the dataset
print("Shape of the lyrics dataset:", lyrics.shape)

# View what the dataset looks like
print(lyrics.head())
print("\n")

# Load the dataset
artists = pd.read_csv('app/artists-data.csv')

# Print the shape of the dataset
print("Shape of the artists dataset:", artists.shape)

# View what the dataset looks like
print(artists.head())
print("\n")

# Merge both datasets for easier processing
singers = pd.merge(lyrics, artists, on='Artist')

# Print the shape of the dataset
print("Shape of the merged dataset:", singers.shape)

# View what the dataset looks like
print(singers.head())
```

```python
# Remove unnecessary white space from the dataset
singers['Lyric'] = singers['Lyric'].str.replace(r'\s+', ' ', regex=True)

```

```python
singers.head()
```

```python
# Filter out songs that have fewer than 20 words
singers = singers[singers['Lyric'].apply(lambda x: len(x.split()) >= 20)]
singers
```

```python

folk_artists = singers[singers['Genres'].str.contains('Folk')]
folk_artists
```

```python


# Find the artist with the highest number of songs among those who sing folk
top_folk_artist = folk_artists[folk_artists['Songs'] == folk_artists['Songs'].max()]

print("Shape of the top folk artist dataset:", top_folk_artist.shape)
print(top_folk_artist.head())
```
