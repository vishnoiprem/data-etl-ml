```python
def is_palindrome(s):
    left = 0
    right = len(s) - 1

    while left < right:

        # skip non-alphanumeric from left
        while left < right and not s[left].isalnum():
            left += 1

        # skip non-alphanumeric from right
        while left < right and not s[right].isalnum():
            right -= 1

        # compare lowercase characters
        if s[left].lower() != s[right].lower():
            return False

        left += 1
        right -= 1

    return True


# input
s = "A  man, a plan, a canal: Panama"

# function call
result = is_palindrome(s)

# output
print(result)
```

```python
def roman_to_int(s):
    values = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000
    }

    total = 0

    for i in range(len(s)):
        current_value = values[s[i]]
        next_value = values[s[i + 1]] if i + 1 < len(s) else 0

        if current_value < next_value:
            total -= current_value
        else:
            total += current_value

    return total


print(roman_to_int("III"))      # 3
print(roman_to_int("LVIII"))    # 58
print(roman_to_int("MCMXCIV"))  # 1994
```

```python
def roman_to_int(s):
    roman = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000
    }

    total = 0
    i = 0

    while i < len(s):
        if i + 1 < len(s) and s[i] == "I" and s[i + 1] == "V":
            total += 4
            i += 2
        elif i + 1 < len(s) and s[i] == "I" and s[i + 1] == "X":
            total += 9
            i += 2
        elif i + 1 < len(s) and s[i] == "X" and s[i + 1] == "L":
            total += 40
            i += 2
        elif i + 1 < len(s) and s[i] == "X" and s[i + 1] == "C":
            total += 90
            i += 2
        elif i + 1 < len(s) and s[i] == "C" and s[i + 1] == "D":
            total += 400
            i += 2
        elif i + 1 < len(s) and s[i] == "C" and s[i + 1] == "M":
            total += 900
            i += 2
        else:
            total += roman[s[i]]
            i += 1

    return total


# input
s = "MCMXCIV"

# function call
result = roman_to_int(s)

# output
print(result)
```

```python
def roman_to_int(s):
    values = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000
    }

    special = {
        "IV": 4,
        "IX": 9,
        "XL": 40,
        "XC": 90,
        "CD": 400,
        "CM": 900
    }

    total = 0
    i = 0

    while i < len(s):
        two_chars = s[i:i + 2]

        if two_chars in special:
            total += special[two_chars]
            i += 2
        else:
            total += values[s[i]]
            i += 1

    return total


# input
s = "MCMXCIV"

# function call
result = roman_to_int(s)

# output
print(result)
```

```python
def intersect(nums1, nums2):
    result = []
    used = [False] * len(nums2)

    for i in range(len(nums1)):
        for j in range(len(nums2)):
            if nums1[i] == nums2[j] and used[j] == False:
                result.append(nums1[i])
                used[j] = True
                break

    return result


# input
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]

# function call
result = intersect(nums1, nums2)

# output
print(result)
```

```python
def intersect(nums1, nums2):
    freq = {}

    for num in nums1:
        freq[num] = freq.get(num, 0) + 1

    result = []

    for num in nums2:
        if num in freq and freq[num] > 0:
            result.append(num)
            freq[num] -= 1

    return result


# input
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]

# function call
result = intersect(nums1, nums2)

# output
print(result)
```

```python
def intersect(nums1, nums2):
    if len(nums1) > len(nums2):
        nums1, nums2 = nums2, nums1

    freq = {}

    for num in nums1:
        freq[num] = freq.get(num, 0) + 1

    result = []

    for num in nums2:
        if freq.get(num, 0) > 0:
            result.append(num)
            freq[num] -= 1

    return result


print(intersect([1, 2, 2, 1], [2, 2]))
print(intersect([4, 9, 5], [9, 4, 9, 8, 4]))
```

```python
def max_profit(prices):
    max_profit_value = 0

    for buy_day in range(len(prices)):
        for sell_day in range(buy_day + 1, len(prices)):
            profit = prices[sell_day] - prices[buy_day]

            if profit > max_profit_value:
                max_profit_value = profit

    return max_profit_value


# input
prices = [7, 1, 5, 3, 6, 4]

# function call
result = max_profit(prices)

# output
print(result)
```
