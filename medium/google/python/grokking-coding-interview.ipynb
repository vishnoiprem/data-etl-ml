```python
# Definition for a Linked List node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


```

```python
def remove_nth_last_node(head, n):
    # 1. Create a dummy node pointing to head to handle edge cases easily
        dummy = ListNode(0, head)

    # 2. Find the total length of the linked list
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # 3. Find the node right BEFORE the one we want to delete
    # The target index from the dummy node is (length - n)
    steps_to_before = length - n
    current = dummy

    for _ in range(steps_to_before):
        current = current.next

    # 4. Skip the N-th node
    current.next = current.next.next

    # Return the real head (not dummy)
    return dummy.next
```

```python


def remove_nth_last_node(head, n):
    if head is None:
        return None

    # First pass: count total nodes
    length = 0
    current = head

    while current is not None:
        length += 1
        current = current.next

    # Optional validation
    if n <= 0 or n > length:
        return head

    # Position from the start, using zero-based index
    target_index = length - n

    # Removing the head node
    if target_index == 0:
        return head.next

    # Second pass: stop at node before target
    current = head

    for _ in range(target_index - 1):
        current = current.next

    # Remove target node
    current.next = current.next.next

    return head
```

```python
def sort_colors(colors):
    for j in range(len(colors)):
        for row in range(j + 1, len(colors)):
            if colors[j] > colors[row]:
                temp = colors[j]
                colors[j] = colors[row]
                colors[row] = temp

    return colors
```

```python
def sort_colors(colors):
    low = 0
    current = 0
    high = len(colors) - 1

    while current <= high:
        if colors[current] == 0:
            colors[low], colors[current] = colors[current], colors[low]
            low += 1
            current += 1

        elif colors[current] == 1:
            current += 1

        else:
            colors[current], colors[high] = colors[high], colors[current]
            high -= 1

    return colors
```

```python
def reverseVowels(s):
    vowels = "aeiouAEIOU"

    s = list(s)   # ✅ convert to list

    start, end = 0, len(s) - 1

    while start < end:

        # move start to vowel
        if s[start] not in vowels:
            start += 1

        # move end to vowel
        elif s[end] not in vowels:
            end -= 1

        # swap vowels
        else:
            s[start], s[end] = s[end], s[start]
            start += 1
            end -= 1

    return "".join(s)
```

```python

def max_sub_array(nums):
    current_sum = nums[0]
    max_sum = nums[0]

    for num in nums[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum
```

```python
def spiralOrder(matrix):
    result = []

    if not matrix:
        return result

    top, bottom = 0, len(matrix) - 1
    left, right = 0, len(matrix[0]) - 1

    while top <= bottom and left <= right:

        # 1. Left → Right
        for j in range(left, right + 1):
            result.append(matrix[top][j])
        top += 1

        # 2. Top → Bottom
        for i in range(top, bottom + 1):
            result.append(matrix[i][right])
        right -= 1

        # 3. Right → Left
        if top <= bottom:
            for j in range(right, left - 1, -1):
                result.append(matrix[bottom][j])
            bottom -= 1

        # 4. Bottom → Top
        if left <= right:
            for i in range(bottom, top - 1, -1):
                result.append(matrix[i][left])
            left += 1

    return result
```

```python
def fib(n, memo={}):
    if n <= 1:
        return n

    if n in memo:
        return memo[n]

    memo[n] = fib(n-1) + fib(n-2)
    return memo[n]
```

```python
def fib(n):
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]
```
