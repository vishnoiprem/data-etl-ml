```python


def max_subarray_len(nums: list[int], k: int) -> int:
    max_len = 0

    for i in range(len(nums)):
        total = 0
        for j in range(i, len(nums)):
            total = total+nums[j]
            if total == k:
                max_len = max(max_len, j - i + 1)

    return max_len

# Test
nums = [1, -1, 5, -2, 3]
k = 3
print(max_subarray_len(nums, k))  # Output: 4
```

```python
def solution_1(nums, k):
    max_len = 0
    n = len(nums)

    for i in range(n):
        for j in range(i, n):
            total = sum(nums[i:j+1])  # recalculate every time
            if total == k:
                max_len = max(max_len, j - i + 1)

    return max_len
```

```python
def solution_2(nums, k):
    max_len = 0
    n = len(nums)

    for i in range(n):
        total = 0
        for j in range(i, n):
            total += nums[j]       # add one element at a time
            if total == k:
                max_len = max(max_len, j - i + 1)

    return max_len
```

```python
def solution_3(nums, k):
    max_len = 0
    n = len(nums)

    # Build prefix sum array
    prefix = [0] * (n + 1)
    for i in range(n):
        prefix[i+1] = prefix[i] + nums[i]

    # prefix[j] - prefix[i] = sum of nums[i..j-1]
    for i in range(n):
        for j in range(i+1, n+1):
            if prefix[j] - prefix[i] == k:
                max_len = max(max_len, j - i)

    return max_len

# prefix for [1,-1,5,-2,3] = [0, 1, 0, 5, 3, 6]
```

```python
def solution_4(nums, k):
    max_len = 0
    prefix_sum = 0
    seen = {0: -1}             # prefix_sum -> earliest index

    for i, num in enumerate(nums):
        prefix_sum = prefix_sum+num

        if prefix_sum - k in seen:
            max_len = max(max_len, i - seen[prefix_sum - k])

        if prefix_sum not in seen:
            seen[prefix_sum] = i   # only store FIRST occurrence

    return max_len
```

```python
def solution_5(nums, k):
    max_len = 0
    cumulative_sum = 0
    sum_index = {}
    sum_index[0] = -1          # empty subarray starts at -1

    for i in range(len(nums)):
        cumulative_sum += nums[i]

        target = cumulative_sum - k

        # check if we've seen (current_sum - k) before
        if target in sum_index:
            length = i - sum_index[target]
            if length > max_len:
                max_len = length

        # record first time we see this cumulative sum
        if cumulative_sum not in sum_index:
            sum_index[cumulative_sum] = i

    return max_len
```
