```python
def spiral_order(matrix):
    if not matrix:
        return []

    rows = len(matrix)
    cols = len(matrix[0])

    visited = [[False] * cols for _ in range(rows)]
    result = []

    directions = [
        (0, 1),   # right
        (1, 0),   # down
        (0, -1),  # left
        (-1, 0)   # up
    ]

    row = 0
    col = 0
    direction_index = 0

    for _ in range(rows * cols):
        result.append(matrix[row][col])
        visited[row][col] = True

        next_row = row + directions[direction_index][0]
        next_col = col + directions[direction_index][1]

        if (
            next_row < 0 or next_row >= rows or
            next_col < 0 or next_col >= cols or
            visited[next_row][next_col]
        ):
            direction_index = (direction_index + 1) % 4
            next_row = row + directions[direction_index][0]
            next_col = col + directions[direction_index][1]

        row = next_row
        col = next_col

    return result


matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(spiral_order(matrix))
```

```python
from collections import Counter, deque
import heapq

def least_interval(tasks, n):
    task_count = Counter(tasks)

    # max heap using negative counts
    heap = []
    for task, count in task_count.items():
        heapq.heappush(heap, (-count, task))

    cooldown = deque()
    time = 0

    while heap or cooldown:
        time += 1

        if heap:
            count, task = heapq.heappop(heap)
            count += 1   # because count is negative

            if count < 0:
                cooldown.append((time + n, count, task))

        if cooldown and cooldown[0][0] == time:
            ready_time, count, task = cooldown.popleft()
            heapq.heappush(heap, (count, task))

    return time


tasks = ["A", "A", "A", "B", "B", "B"]
n = 2

print(least_interval(tasks, n))
```

```python
class Codec:
    def encode(self, strs):
        parts = []

        for word in strs:
            parts.append(str(len(word)))
            parts.append("#")
            parts.append(word)

        return "".join(parts)

    def decode(self, s):
        result = []
        i = 0

        while i < len(s):
            j = i

            while s[j] != "#":
                j += 1

            length = int(s[i:j])

            start = j + 1
            end = start + length

            result.append(s[start:end])

            i = end

        return result


codec = Codec()

strs = ["hello", "wo#rld", "", "abc"]
encoded = codec.encode(strs)
decoded = codec.decode(encoded)

print(encoded)
print(decoded)
```
