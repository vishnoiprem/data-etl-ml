```python
import csv
import heapq

def create_sample_csv(file_path):
    rows = [
        {"user_id": "1", "name": "Amit", "revenue": "500"},
        {"user_id": "2", "name": "Prem", "revenue": "900"},
        {"user_id": "3", "name": "Sara", "revenue": "700"},
        {"user_id": "4", "name": "John", "revenue": "300"},
        {"user_id": "5", "name": "Nina", "revenue": "1200"},
    ]

    with open(file_path, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=["user_id", "name", "revenue"])
        writer.writeheader()
        writer.writerows(rows)


def top_n_by_column(file_path, column_name, n):
    heap = []

    with open(file_path, mode="r", newline="", encoding="utf-8") as file:
        reader = csv.DictReader(file)

        for row_number, row in enumerate(reader, start=2):
            try:
                value = float(row[column_name])
            except ValueError:
                print(f"Skipping bad row {row_number}: invalid number")
                continue
            except KeyError:
                raise KeyError(f"Column '{column_name}' not found in CSV")

            row[column_name] = value

            heapq.heappush(heap, (value, row))

            if len(heap) > n:
                heapq.heappop(heap)

    result = [item[1] for item in heap]
    result.sort(key=lambda x: x[column_name], reverse=True)

    return result


# input
file_path = "sales.csv"
create_sample_csv(file_path)

# function call
top_rows = top_n_by_column(file_path, "revenue", 3)

# output
for row in top_rows:
    print(row)
```

```python
def flatten_dict(data, parent_key=""):
    result = {}

    for key, value in data.items():
        print(key,value,'kv')
        if parent_key:
            new_key = parent_key + "." + key
        else:
            new_key = key

        if isinstance(value, dict):
            nested_result = flatten_dict(value, new_key)
            result.update(nested_result)
        else:
            result[new_key] = value

    return result


data = {
    "user": {
        "id": 1,
        "name": "Prem",
        "address": {
            "city": "Bangkok",
            "country": "Thailand"
        }
    },
    "event": "login"
}

print(flatten_dict(data))
```

```python
def flatten_json(data, parent_key=""):
    result = {}

    if isinstance(data, dict):
        for key, value in data.items():
            if parent_key:
                new_key = parent_key + "." + str(key)
            else:
                new_key = str(key)

            result.update(flatten_json(value, new_key))

    elif isinstance(data, list):
        for index, value in enumerate(data):
            if parent_key:
                new_key = parent_key + "." + str(index)
            else:
                new_key = str(index)

            result.update(flatten_json(value, new_key))

    else:
        result[parent_key] = data

    return result


data = {
    "user": {
        "id": 1,
        "name": "Prem",
        "orders": [
            {"order_id": "O1", "amount": 100},
            {"order_id": "O2", "amount": 200}
        ]
    },
    "event": "purchase"
}

print(flatten_json(data))
```

```python
def flatten(nested):
    result = []
    for item in nested:
        if isinstance(item, list):
            result.extend(flatten(item))   # item is a list → recurse
        else:
            result.append(item)            # item is a value → keep it
    return result
```
