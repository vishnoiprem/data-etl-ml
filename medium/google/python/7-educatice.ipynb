```python
def three_sum(nums):
    nums.sort()
    result = []
    n = len(nums)

    for i in range(n - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        if nums[i] > 0:
            break
        left = i + 1
        right = n - 1
        while left < right:
            currentSum = nums[i] + nums[left] + nums[right]
            if currentSum < 0:
                left += 1
            elif currentSum > 0:
                right -= 1
            else:
                result.append([nums[i], nums[left], nums[right]])
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1

    return result


def main():
    testCases = [
        ([-2, 0, 1, 1, 2],),
        ([1, -1, -1, 0],),
        ([-4, -2, -1, 0, 1, 2, 3, 4],),
        ([3, -3, 0, -1, 1, 2, -2],),
        ([-5, -3, -1, 0, 1, 2, 3, 5],),
    ]

    for idx, (nums,) in enumerate(testCases, 1):
        result = three_sum(nums)
        print(f"{idx}.\tInput array: {nums}")
        print(f"\tResult: {result}")
        print("-" * 100)


if __name__ == "__main__":
    main()
```
