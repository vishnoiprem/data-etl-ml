## 420. Long Pressed Name

**Topic:** Algorithms  
**Difficulty:** Easy

Check whether `typed` can be formed from `name` by repeating some characters.

**Pattern:** two pointers with a controlled skip rule.

### Simple → Better → Best

Simple: simulate; Better: run-length encoding; Best: two pointers

### Time and space

O(n+m) time, O(1) extra space for best

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def long_pressed_simple(name, typed):
    i = j = 0
    while i < len(name) and j < len(typed):
        if name[i] == typed[j]:
            i += 1; j += 1
        elif j > 0 and typed[j] == typed[j-1]:
            j += 1
        else:
            return False
    while j < len(typed) and typed[j] == typed[j-1]:
        j += 1
    return i == len(name) and j == len(typed)

def long_pressed_grouped(name, typed):
    def groups(s):
        out, i = [], 0
        while i < len(s):
            j = i
            while j < len(s) and s[j] == s[i]:
                j += 1
            out.append((s[i], j-i))
            i = j
        return out
    a, b = groups(name), groups(typed)
    return len(a) == len(b) and all(c1 == c2 and n1 <= n2 for (c1,n1),(c2,n2) in zip(a,b))

print(long_pressed_simple("alex", "aaleex"))
```

## 422. 01 Matrix

**Topic:** Data Structures  
**Difficulty:** Medium

For each cell, find distance to the nearest zero.

**Pattern:** multi-source BFS.

### Simple → Better → Best

Simple: nearest-zero scan; Better: DP two passes; Best: multi-source BFS

### Time and space

O(mn) time, O(mn) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def update_matrix_bruteforce(mat):
    rows, cols = len(mat), len(mat[0])
    out = [[0]*cols for _ in range(rows)]
    zeros = [(r,c) for r in range(rows) for c in range(cols) if mat[r][c] == 0]
    for r in range(rows):
        for c in range(cols):
            out[r][c] = min(abs(r-zr)+abs(c-zc) for zr,zc in zeros)
    return out

def update_matrix_bfs(mat):
    rows, cols = len(mat), len(mat[0])
    dist = [[-1]*cols for _ in range(rows)]
    q = deque()
    for r in range(rows):
        for c in range(cols):
            if mat[r][c] == 0:
                dist[r][c] = 0
                q.append((r,c))
    for r,c in iter(q):
        pass
    while q:
        r,c = q.popleft()
        for dr,dc in ((1,0),(-1,0),(0,1),(0,-1)):
            nr,nc = r+dr,c+dc
            if 0 <= nr < rows and 0 <= nc < cols and dist[nr][nc] == -1:
                dist[nr][nc] = dist[r][c] + 1
                q.append((nr,nc))
    return dist

mat = [[0,0,0],[0,1,0],[1,1,1]]
print(update_matrix_bfs(mat))
```

## 424. Valid Perfect Square

**Topic:** Math  
**Difficulty:** Easy

Determine whether an integer is a perfect square without using `sqrt`.

**Pattern:** binary search on a monotonic condition.

### Simple → Better → Best

Simple: linear scan; Better: Newton's method; Best: binary search

### Time and space

Best: O(log n) time, O(1) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def is_square_linear(n):
    i = 0
    while i*i < n:
        i += 1
    return i*i == n

def is_square_newton(n):
    if n < 2: return True
    x = n
    while x*x > n:
        x = (x + n//x)//2
    return x*x == n

def is_square_binary(n):
    left, right = 0, n
    while left <= right:
        mid = (left + right)//2
        sq = mid*mid
        if sq == n: return True
        if sq < n: left = mid+1
        else: right = mid-1
    return False

for n in [16, 14, 1, 0]:
    print(n, is_square_binary(n))
```

## 428. Interval List Intersections

**Topic:** Data Structures  
**Difficulty:** Medium

Find intersections between two sorted lists of disjoint intervals.

**Pattern:** two pointers; whichever interval ends first cannot intersect future intervals.

### Simple → Better → Best

Simple: all pairs; Better: merge-style scan; Best: two pointers

### Time and space

O(m+n) time, O(1) extra space excluding output

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def intersections_bruteforce(a, b):
    out = []
    for x1, x2 in a:
        for y1, y2 in b:
            lo, hi = max(x1,y1), min(x2,y2)
            if lo <= hi:
                out.append([lo,hi])
    return out

def intersections_two_pointers(a, b):
    i = j = 0
    out = []
    while i < len(a) and j < len(b):
        lo = max(a[i][0], b[j][0])
        hi = min(a[i][1], b[j][1])
        if lo <= hi:
            out.append([lo,hi])
        if a[i][1] < b[j][1]:
            i += 1
        else:
            j += 1
    return out

A = [[0,2],[5,10],[13,23],[24,25]]
B = [[1,5],[8,12],[15,24],[25,26]]
print(intersections_two_pointers(A,B))
```

## 464. Top 10 Highest-Rated Hotels

**Topic:** Database  
**Difficulty:** Medium

Deduplicate repeated hotel-review rows, rank scores descending, and include ties at the cutoff.

### Simple → Better → Best

Simple: DISTINCT + LIMIT; Better: find cutoff score; Best: RANK after one row per hotel

### Time and space

Grouping and sorting dominate

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

sql = '''
WITH hotel_scores AS (
    SELECT hotel_name, MAX(average_score) AS average_score
    FROM hotel_reviews
    GROUP BY hotel_name
),
ranked AS (
    SELECT
        hotel_name,
        average_score,
        RANK() OVER (ORDER BY average_score DESC) AS score_rank
    FROM hotel_scores
)
SELECT hotel_name, average_score
FROM ranked
WHERE score_rank <= 10
ORDER BY average_score DESC, hotel_name;
'''
print(sql)
```

## 475. Minimum Path Sum

**Topic:** Algorithms  
**Difficulty:** Medium

Find the minimum path sum from top-left to bottom-right moving right/down.

**Pattern:** grid dynamic programming.

### Simple → Better → Best

Simple: recursion; Better: 2D DP; Best: 1D rolling DP

### Time and space

Best: O(mn) time, O(n) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def min_path_recursive(grid):
    rows, cols = len(grid), len(grid[0])
    def f(r,c):
        if r == rows-1 and c == cols-1: return grid[r][c]
        if r >= rows or c >= cols: return float('inf')
        return grid[r][c] + min(f(r+1,c), f(r,c+1))
    return f(0,0)

def min_path_dp(grid):
    rows, cols = len(grid), len(grid[0])
    dp = [[0]*cols for _ in range(rows)]
    for r in range(rows-1,-1,-1):
        for c in range(cols-1,-1,-1):
            if r == rows-1 and c == cols-1:
                dp[r][c] = grid[r][c]
            else:
                down = dp[r+1][c] if r+1 < rows else float('inf')
                right = dp[r][c+1] if c+1 < cols else float('inf')
                dp[r][c] = grid[r][c] + min(down,right)
    return dp[0][0]

def min_path_1d(grid):
    cols = len(grid[0])
    dp = [float('inf')]*cols
    dp[0] = 0
    for row in grid:
        for c, value in enumerate(row):
            if c == 0:
                dp[c] = dp[c] + value
            else:
                dp[c] = min(dp[c], dp[c-1]) + value
    return dp[-1]

grid = [[1,3,1],[1,5,1],[4,2,1]]
print(min_path_dp(grid), min_path_1d(grid))
```

```python

def is_subsequence_simple(s,t):
    i = 0
    for ch in t:
        if i < len(s) and s[i] == ch:
            i += 1
    return i == len(s)

def is_subsequence_indexed(s,t):
    pos = defaultdict(list)
    for i,ch in enumerate(t):
        pos[ch].append(i)
    current = -1
    import bisect
    for ch in s:
        arr = pos[ch]
        j = bisect.bisect_right(arr,current)
        if j == len(arr): return False
        current = arr[j]
    return True

print(is_subsequence_simple("abc","ahbgdc"))
```

## 568. Squares of a Sorted Array

**Topic:** Algorithms  
**Difficulty:** Easy

Square a sorted array and return squares in sorted order.

**Pattern:** largest absolute value is at one of the ends.

### Simple → Better → Best

Simple: square + sort; Better: split and merge; Best: two pointers

### Time and space

Best: O(n) time, O(n) output space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def squares_simple(nums):
    return sorted(x*x for x in nums)

def squares_merge(nums):
    neg = [x*x for x in nums if x < 0][::-1]
    pos = [x*x for x in nums if x >= 0]
    i = j = 0
    out = []
    while i < len(neg) and j < len(pos):
        if neg[i] <= pos[j]:
            out.append(neg[i]); i += 1
        else:
            out.append(pos[j]); j += 1
    return out + neg[i:] + pos[j:]

def squares_best(nums):
    out = [0]*len(nums)
    left, right = 0, len(nums)-1
    write = len(nums)-1
    while left <= right:
        l, r = nums[left]**2, nums[right]**2
        if l > r:
            out[write] = l; left += 1
        else:
            out[write] = r; right -= 1
        write -= 1
    return out

print(squares_best([-4,-1,0,3,10]))
```

## 573. Reorganize String

**Topic:** Data Structures  
**Difficulty:** Medium

Rearrange characters so adjacent characters differ.

**Pattern:** greedily use the two most frequent available characters.

### Simple → Better → Best

Simple: backtracking; Better: sorted frequency placement; Best: max heap

### Time and space

O(n log a) time, O(a) space where a is alphabet size

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def reorganize_heap(s):
    counts = Counter(s)
    if max(counts.values()) > (len(s)+1)//2:
        return ""
    heap = [(-count,ch) for ch,count in counts.items()]
    heapq.heapify(heap)
    out = []
    prev_count, prev_ch = 0, ''
    while heap:
        count,ch = heapq.heappop(heap)
        out.append(ch)
        count += 1
        if prev_count < 0:
            heapq.heappush(heap,(prev_count,prev_ch))
        prev_count,prev_ch = count,ch
    return ''.join(out)

print(reorganize_heap("aab"))
```

## 604. F1 Score

**Topic:** ML Coding  
**Difficulty:** Medium

Combine precision and recall using their harmonic mean.

### Simple → Better → Best

Simple: confusion matrix counts; Better: vectorized; Best: macro/micro/weighted-aware implementation

### Time and space

O(n) time, O(1) extra space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def f1_from_labels(y_true, y_pred):
    tp = fp = fn = 0
    for y,p in zip(y_true,y_pred):
        tp += (y == 1 and p == 1)
        fp += (y == 0 and p == 1)
        fn += (y == 1 and p == 0)
    precision = tp/(tp+fp) if tp+fp else 0.0
    recall = tp/(tp+fn) if tp+fn else 0.0
    return 2*precision*recall/(precision+recall) if precision+recall else 0.0

print(f1_from_labels([1,1,0,1],[1,0,1,1]))
```

## 627. Find First and Last Position

**Topic:** Algorithms  
**Difficulty:** Medium

Find the first and last occurrence of a target in a sorted array.

**Pattern:** run binary search twice with different boundary rules.

### Simple → Better → Best

Simple: scan; Better: one binary + expand; Best: two boundary binary searches

### Time and space

O(log n) time, O(1) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def first_last_linear(nums, target):
    idx = [i for i,x in enumerate(nums) if x == target]
    return [-1,-1] if not idx else [idx[0],idx[-1]]

def lower_bound(nums, target):
    left,right = 0,len(nums)
    while left < right:
        mid = (left+right)//2
        if nums[mid] < target: left = mid+1
        else: right = mid
    return left

def first_last_binary(nums, target):
    first = lower_bound(nums,target)
    if first == len(nums) or nums[first] != target:
        return [-1,-1]
    last = lower_bound(nums,target+1)-1
    return [first,last]

print(first_last_binary([5,7,7,8,8,10],8))
```

## 633. Population Variance

**Topic:** Statistics  
**Difficulty:** Easy

Compute average squared deviation from the population mean.

### Simple → Better → Best

### Time and space

O(n) time, O(1) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def variance_two_pass(values):
    mean = sum(values)/len(values)
    return sum((x-mean)**2 for x in values)/len(values)

def variance_formula(values):
    mean = sum(values)/len(values)
    return sum(x*x for x in values)/len(values) - mean*mean

def variance_welford(values):
    n = 0
    mean = m2 = 0.0
    for x in values:
        n += 1
        delta = x-mean
        mean += delta/n
        m2 += delta*(x-mean)
    return m2/n

data=[2,4,4,4,5,5,7,9]
print(variance_two_pass(data), variance_welford(data))
```

## 648. Evaluate Reverse Polish Notation

**Topic:** Data Structures  
**Difficulty:** Medium

Evaluate postfix arithmetic.

**Pattern:** operands go on a stack; an operator consumes the top two values.

### Simple → Better → Best

Simple: repeatedly reduce expression; Better: recursive parser; Best: stack

### Time and space

O(n) time, O(n) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def eval_rpn(tokens):
    stack = []
    for token in tokens:
        if token in {"+","-","*","/"}:
            b = stack.pop(); a = stack.pop()
            if token == "+": stack.append(a+b)
            elif token == "-": stack.append(a-b)
            elif token == "*": stack.append(a*b)
            else: stack.append(int(a/b))
        else:
            stack.append(int(token))
    return stack[-1]

print(eval_rpn(["2","1","+","3","*"]))
```

## 657. Partition Labels

**Topic:** Algorithms  
**Difficulty:** Medium

Split a string so each character appears in at most one part.

**Pattern:** track each character's last occurrence and close a partition when current index reaches the farthest required end.

### Simple → Better → Best

Simple: test every split; Better: interval merge; Best: last-index greedy

### Time and space

O(n) time, O(a) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def partition_labels(s):
    last = {ch:i for i,ch in enumerate(s)}
    start = end = 0
    out = []
    for i,ch in enumerate(s):
        end = max(end,last[ch])
        if i == end:
            out.append(end-start+1)
            start = i+1
    return out

print(partition_labels("ababcbacadefegdehijhklij"))
```

## 681. Longest Consecutive Sequence

**Topic:** Data Structures  
**Difficulty:** Medium

Find the longest run of consecutive integers in an unsorted array.

**Pattern:** hash set and only start counting at sequence starts.

### Simple → Better → Best

Simple: sort; Better: union-find; Best: hash-set sequence starts

### Time and space

Best: O(n) average time, O(n) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def longest_sort(nums):
    if not nums: return 0
    nums = sorted(set(nums))
    best = cur = 1
    for i in range(1,len(nums)):
        if nums[i] == nums[i-1]+1:
            cur += 1
        else:
            best = max(best,cur); cur = 1
    return max(best,cur)

def longest_set(nums):
    values = set(nums)
    best = 0
    for x in values:
        if x-1 not in values:
            y = x
            while y in values:
                y += 1
            best = max(best, y-x)
    return best

print(longest_set([100,4,200,1,3,2]))
```

## 683. Minimum Number of Days to Make m Bouquets

**Topic:** Algorithms  
**Difficulty:** Medium

Find the earliest day when at least `m` bouquets of `k` adjacent flowers can be made.

**Pattern:** binary search on the answer.

### Simple → Better → Best

Simple: test every day; Better: test sorted unique days; Best: binary search answer

### Time and space

O(n log range) time, O(1) space

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def min_days(bloom, m, k):
    if m*k > len(bloom):
        return -1
    def can(day):
        bouquets = flowers = 0
        for b in bloom:
            if b <= day:
                flowers += 1
                if flowers == k:
                    bouquets += 1
                    flowers = 0
            else:
                flowers = 0
        return bouquets >= m
    left,right = min(bloom),max(bloom)
    while left < right:
        mid = (left+right)//2
        if can(mid): right = mid
        else: left = mid+1
    return left

print(min_days([1,10,3,10,2],3,1))
```

## 688. Minimum Time Difference

**Topic:** Algorithms  
**Difficulty:** Medium

Find the minimum circular difference between times in a day.

**Pattern:** normalize to minutes, sort, and check adjacent plus wrap-around.

### Simple → Better → Best

Simple: all pairs; Better: sort; Best: boolean bucket of 1440 minutes

### Time and space

Sort: O(n log n); bucket: O(n+1440)

### What to say in the interview

> “First I will clarify the input, output, and edge cases.  
> My simple solution is ____, with time ____ and space ____.  
> The repeated work is ____. I can improve it using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main edge case.”


```python

def min_time_difference(times):
    minutes = []
    for t in times:
        h,m = map(int,t.split(':'))
        minutes.append(h*60+m)
    minutes.sort()
    best = 1440 - minutes[-1] + minutes[0]
    for a,b in zip(minutes,minutes[1:]):
        best = min(best,b-a)
    return best

print(min_time_difference(["23:59","00:00"]))
```

## Pattern map

- Sorted array → two pointers or binary search
- Fixed window → sliding window
- Nested structure → stack or recursion
- Shortest path in unweighted graph → BFS

- Top K stream → heap
- SQL ranking → aggregate first, then window function
- No matching row → NOT EXISTS / anti-join
- Streaming → event time, watermark, state, idempotency
