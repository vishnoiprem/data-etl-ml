## 1112. Hamming Distance

**Topic:** Algorithms  
**Difficulty:** Easy

Count bit positions where two integers differ.

**How to think:** XOR produces `1` exactly where the bits differ.

### Simple → Better → Best

Simple: binary string count; Better: inspect every bit; Best: Kernighan removes one set bit at a time

### Time and space

Best: O(number of different bits), O(1) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def hamming_string(x, y):
    return bin(x ^ y).count("1")

def hamming_loop(x, y):
    value = x ^ y
    count = 0
    while value:
        count += value & 1
        value >>= 1
    return count

def hamming_kernighan(x, y):
    value = x ^ y
    count = 0
    while value:
        value &= value - 1
        count += 1
    return count

print(hamming_string(1, 4), hamming_kernighan(1, 4))
```

## 1118. Longest Repeating Character Replacement

**Topic:** Algorithms  
**Difficulty:** Medium

Find the longest substring that can be made of one repeated character after at most `k` replacements.

**Pattern:** variable sliding window.

### Simple → Better → Best

Simple: enumerate substrings; Better: recompute max frequency per window; Best: sliding window with monotonic max frequency

### Time and space

Best: O(n) time, O(alphabet) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def character_replacement_bruteforce(s, k):
    best = 0
    for left in range(len(s)):
        counts = Counter()
        for right in range(left, len(s)):
            counts[s[right]] += 1
            window = right-left+1
            if window - max(counts.values()) <= k:
                best = max(best, window)
    return best

def character_replacement_best(s, k):
    counts = Counter()
    left = best = max_freq = 0
    for right, ch in enumerate(s):
        counts[ch] += 1
        max_freq = max(max_freq, counts[ch])
        while (right-left+1) - max_freq > k:
            counts[s[left]] -= 1
            left += 1
        best = max(best, right-left+1)
    return best

print(character_replacement_best("AABABBA", 1))
```

## 1120. Find the Difference

**Topic:** Algorithms  
**Difficulty:** Easy

One string contains all characters of another plus one extra character.

### Simple → Better → Best

Simple: sort; Better: frequency map; Best: XOR

### Time and space

Best: O(n) time, O(1) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def find_diff_sort(s, t):
    for a,b in zip(sorted(s), sorted(t)):
        if a != b: return b
    return sorted(t)[-1]

def find_diff_count(s, t):
    counts = Counter(s)
    for ch in t:
        counts[ch] -= 1
        if counts[ch] < 0:
            return ch

def find_diff_xor(s, t):
    value = 0
    for ch in s+t:
        value ^= ord(ch)
    return chr(value)

print(find_diff_xor("abcd", "abcde"))
```

## 1136. Combination Sum IV

**Topic:** Algorithms  
**Difficulty:** Medium

Count ordered sequences that sum to a target.

**Pattern:** dynamic programming where order matters.

### Simple → Better → Best

Simple: recursion; Better: memoization; Best: bottom-up DP

### Time and space

O(target × len(nums)) time, O(target) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def combination_sum_recursive(nums, target):
    if target == 0: return 1
    if target < 0: return 0
    return sum(combination_sum_recursive(nums, target-x) for x in nums)

def combination_sum_memo(nums, target):
    memo = {0: 1}
    def f(rem):
        if rem < 0: return 0
        if rem not in memo:
            memo[rem] = sum(f(rem-x) for x in nums)
        return memo[rem]
    return f(target)

def combination_sum_dp(nums, target):
    dp = [0]*(target+1)
    dp[0] = 1
    for total in range(1, target+1):
        for x in nums:
            if x <= total:
                dp[total] += dp[total-x]
    return dp[target]

print(combination_sum_dp([1,2,3], 4))
```

## 1139. Minimum Operations to Reduce X to Zero

**Topic:** Algorithms  
**Difficulty:** Medium

Remove values from either end so their sum equals `x`.

**How to think:** keep the longest middle subarray whose sum is `total-x`.

### Simple → Better → Best

Simple: try left/right removals; Better: prefix + suffix lookup; Best: longest middle sliding window

### Time and space

Best: O(n) time, O(1) space for positive nums

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def min_ops_bruteforce(nums, x):
    n = len(nums)
    best = float("inf")
    prefix = [0]
    for v in nums:
        prefix.append(prefix[-1]+v)
    suffix = [0]
    for v in reversed(nums):
        suffix.append(suffix[-1]+v)
    for left in range(n+1):
        for right in range(n-left+1):
            if prefix[left] + suffix[right] == x:
                best = min(best, left+right)
    return -1 if best == float("inf") else best

def min_ops_sliding(nums, x):
    target = sum(nums)-x
    if target < 0: return -1
    left = current = 0
    longest = -1
    for right, v in enumerate(nums):
        current += v
        while current > target and left <= right:
            current -= nums[left]
            left += 1
        if current == target:
            longest = max(longest, right-left+1)
    return -1 if longest == -1 else len(nums)-longest

print(min_ops_sliding([1,1,4,2,3], 5))
```

## 1170. Koko Eating Bananas

**Topic:** Algorithms  
**Difficulty:** Medium

Find the minimum eating speed that finishes all piles within `h` hours.

**Pattern:** binary search on the answer.

### Simple → Better → Best

Simple: test every speed; Better: search unique candidates; Best: binary search speed

### Time and space

O(n log max(pile)) time, O(1) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def min_eating_speed(piles, h):
    def hours(speed):
        return sum((pile + speed - 1)//speed for pile in piles)
    left, right = 1, max(piles)
    while left < right:
        mid = (left+right)//2
        if hours(mid) <= h:
            right = mid
        else:
            left = mid+1
    return left

print(min_eating_speed([3,6,7,11], 8))
```

> I will test a normal case, the smallest case, and the main boundary case.”

### What to say in the interview

> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def k_closest_sort(arr, k, x):
    return sorted(sorted(arr, key=lambda v: (abs(v-x), v))[:k])

def k_closest_heap(arr, k, x):
    heap = []
    for v in arr:
        heapq.heappush(heap, (-abs(v-x), -v))
        if len(heap) > k:
            heapq.heappop(heap)
    return sorted(-v for _,v in heap)

def k_closest_binary(arr, k, x):
    left, right = 0, len(arr)-k
    while left < right:
        mid = (left+right)//2
        if x-arr[mid] > arr[mid+k]-x:
            left = mid+1
        else:
            right = mid
    return arr[left:left+k]

print(k_closest_binary([1,2,3,4,5], 4, 3))
```

## 1216. Flood Fill

**Topic:** Data Structures  
**Difficulty:** Easy

Recolor the connected component containing the start cell.

**Pattern:** DFS/BFS over four neighbors.

### Simple → Better → Best

Simple: scan repeatedly; Better: DFS; Best: BFS/DFS with visited-by-recolor

### Time and space

O(mn) time, O(mn) worst-case stack/queue

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def flood_fill_dfs(image, sr, sc, color):
    old = image[sr][sc]
    if old == color: return image
    rows, cols = len(image), len(image[0])
    def dfs(r,c):
        if not (0 <= r < rows and 0 <= c < cols) or image[r][c] != old:
            return
        image[r][c] = color
        for dr,dc in ((1,0),(-1,0),(0,1),(0,-1)):
            dfs(r+dr,c+dc)
    dfs(sr,sc)
    return image

img = [[1,1,1],[1,1,0],[1,0,1]]
print(flood_fill_dfs(img,1,1,2))
```

**Difficulty:** Easy

Check whether any value appears more than once.

### Simple → Better → Best

Simple: nested loops; Better: sort; Best: hash set

### Time and space

Best: O(n) average time, O(n) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.

> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def contains_duplicate_sort(nums):
    nums = sorted(nums)
    return any(nums[i] == nums[i-1] for i in range(1,len(nums)))

def contains_duplicate_count(nums):
    return any(v > 1 for v in Counter(nums).values())

def contains_duplicate_set(nums):
    seen = set()
    for x in nums:
        if x in seen: return True
        seen.add(x)
    return False

print(contains_duplicate_set([1,2,3,1]))
```

```python

def lcp_vertical(words):
    if not words: return ""
    for i,ch in enumerate(words[0]):
        for word in words[1:]:
            if i == len(word) or word[i] != ch:
                return words[0][:i]
    return words[0]

def lcp_sort(words):
    if not words: return ""
    first,last = min(words),max(words)
    i = 0
    while i < min(len(first),len(last)) and first[i] == last[i]:
        i += 1
    return first[:i]

print(lcp_vertical(["flower","flow","flight"]))
```

**Topic:** Algorithms  
**Difficulty:** Medium

Count subarrays containing exactly `k` odd numbers.

**Pattern:** prefix counts or `atMost(k)-atMost(k-1)`.

### Simple → Better → Best

### Time and space

O(n) time, O(n) or O(1) space

### What to say in the interview

> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def nice_prefix(nums, k):
    counts = {0:1}
    odd = result = 0
    for x in nums:
        odd += x % 2
        result += counts.get(odd-k,0)
        counts[odd] = counts.get(odd,0)+1
    return result

def nice_at_most(nums, k):
    def at_most(limit):
        left = result = 0
        for right,x in enumerate(nums):
            limit -= x%2
            while limit < 0:
                limit += nums[left]%2
                left += 1
            result += right-left+1
        return result
    return at_most(k)-at_most(k-1)

print(nice_prefix([1,1,2,1,1],3))
```

```python

def min_arrows(points):
    if not points: return 0
    points = sorted(points, key=lambda p:p[1])
    arrows = 1
    shot = points[0][1]
    for start,end in points[1:]:
        if start > shot:
            arrows += 1
            shot = end
    return arrows

print(min_arrows([[10,16],[2,8],[1,6],[7,12]]))
```

## 1298. Longest Palindromic Substring

**Topic:** Algorithms  
**Difficulty:** Medium

Find the longest contiguous palindrome.

**Pattern:** every palindrome expands from a center.

### Simple → Better → Best

Simple: all substrings; Better: DP; Best: expand around centers

### Time and space

Best: O(n²) time, O(1) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def longest_pal_bruteforce(s):
    best = ""
    for i in range(len(s)):
        for j in range(i+1,len(s)+1):
            part = s[i:j]
            if part == part[::-1] and len(part) > len(best):
                best = part
    return best

def longest_pal_center(s):
    best_l = best_r = 0
    def expand(l,r):
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l -= 1; r += 1
        return l+1,r-1
    for i in range(len(s)):
        for l,r in (expand(i,i), expand(i,i+1)):
            if r-l > best_r-best_l:
                best_l,best_r = l,r
    return s[best_l:best_r+1]

print(longest_pal_center("babad"))
```

```python

def car_fleet(target, position, speed):
    cars = sorted(zip(position,speed), reverse=True)
    fleets = 0
    last_time = -1
    for pos,spd in cars:
        time = (target-pos)/spd
        if time > last_time:
            fleets += 1
            last_time = time
    return fleets

print(car_fleet(12,[10,8,0,5,3],[2,4,1,1,3]))
```

### Simple → Better → Best

### What to say in the interview

> “I will first clarify the input, output, and edge cases.

> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def rotate_extra(nums,k):
    k %= len(nums)
    return nums[-k:]+nums[:-k]

def rotate_repeated(nums,k):
    for _ in range(k%len(nums)):
        nums.insert(0,nums.pop())
    return nums

def rotate_reverse(nums,k):
    n = len(nums); k %= n
    def rev(l,r):
        while l < r:
            nums[l],nums[r] = nums[r],nums[l]
            l += 1; r -= 1
    rev(0,n-1); rev(0,k-1); rev(k,n-1)
    return nums

print(rotate_reverse([1,2,3,4,5,6,7],3))
```

## 1365. Happy Number

**Topic:** Math  
**Difficulty:** Easy

Repeatedly replace a number by the sum of squares of its digits; determine whether it reaches 1.

**Pattern:** cycle detection.

### Simple → Better → Best

Simple: bounded simulation; Better: set cycle detection; Best: Floyd cycle detection

### Time and space

O(log n) per transformation, O(1) extra space for Floyd

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def next_happy(n):
    total = 0
    while n:
        n,d = divmod(n,10)
        total += d*d
    return total

def happy_set(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = next_happy(n)
    return n == 1

def happy_floyd(n):
    slow = fast = n
    while True:
        slow = next_happy(slow)
        fast = next_happy(next_happy(fast))
        if slow == fast:
            return slow == 1

print(happy_floyd(19))
```

## 1376. Coin Change II

**Topic:** Algorithms  
**Difficulty:** Medium

Count combinations of coins that make an amount.

**Key detail:** loop coins outside and amounts inside so order does not create duplicates.

### Simple → Better → Best

Simple: recursion; Better: memoization; Best: 1D combination DP

### Time and space

O(amount × coins) time, O(amount) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def change_recursive(amount, coins):
    def f(i, rem):
        if rem == 0: return 1
        if rem < 0 or i == len(coins): return 0
        return f(i, rem-coins[i]) + f(i+1, rem)
    return f(0,amount)

def change_dp(amount, coins):
    dp = [0]*(amount+1)
    dp[0] = 1
    for coin in coins:
        for total in range(coin,amount+1):
            dp[total] += dp[total-coin]
    return dp[amount]

print(change_dp(5,[1,2,5]))
```

## 1460. Merge Intervals

**Topic:** Data Structures  
**Difficulty:** Medium

Merge all overlapping intervals.

**Pattern:** sort by start, then compare against the last merged interval.

### Simple → Better → Best

Simple: repeatedly merge pairs; Better: graph components; Best: sort and sweep

### Time and space

O(n log n) time, O(n) output

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def merge_intervals(intervals):
    intervals = sorted(intervals)
    merged = []
    for start,end in intervals:
        if not merged or start > merged[-1][1]:
            merged.append([start,end])
        else:
            merged[-1][1] = max(merged[-1][1],end)
    return merged

print(merge_intervals([[1,3],[2,6],[8,10],[15,18]]))
```

## 1465. Maximize Sum of Array After K Negations

**Topic:** Algorithms  
**Difficulty:** Easy

Negate values exactly `k` times to maximize sum.

**Pattern:** flip the most negative values first; if one odd flip remains, flip the smallest absolute value.

### Simple → Better → Best

Simple: search all choices; Better: min-heap repeated flips; Best: greedy negatives + parity

### Time and space

O(n log n) time, O(1) extra after sorting

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def max_sum_k_negations(nums,k):
    nums.sort()
    i = 0
    while i < len(nums) and k > 0 and nums[i] < 0:
        nums[i] = -nums[i]
        i += 1
        k -= 1
    total = sum(nums)
    if k % 2:
        total -= 2*min(abs(x) for x in nums)
    return total

print(max_sum_k_negations([4,2,3],1))
```

## 1468. Unique Paths

**Topic:** Algorithms  
**Difficulty:** Medium

Count paths from top-left to bottom-right moving only right and down.

### Simple → Better → Best

Simple: recursion; Better: 2D DP; Best: 1D rolling DP or combinatorics

### Time and space

Best DP: O(mn) time, O(n) space

### What to say in the interview

> “I will first clarify the input, output, and edge cases.  
> The simple solution is ____, with time ____ and space ____.  
> Its repeated work is ____. I can improve that using ____.  
> The key invariant/state is ____.  
> I will test a normal case, the smallest case, and the main boundary case.”


```python

def unique_recursive(m,n):
    if m == 1 or n == 1: return 1
    return unique_recursive(m-1,n)+unique_recursive(m,n-1)

def unique_dp(m,n):
    dp = [[1]*n for _ in range(m)]
    for r in range(1,m):
        for c in range(1,n):
            dp[r][c] = dp[r-1][c]+dp[r][c-1]
    return dp[-1][-1]

def unique_1d(m,n):
    dp = [1]*n
    for _ in range(1,m):
        for c in range(1,n):
            dp[c] += dp[c-1]
    return dp[-1]

print(unique_dp(3,7), unique_1d(3,7))
```
