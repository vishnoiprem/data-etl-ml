## 1504. Cherry Pickup II

**Topic:** Algorithms  
**Difficulty:** Hard

Two robots move row by row and collect cherries, counting a shared cell once.

**Pattern:** DP state is `(row, col1, col2)`.

### Simple → Better → Best

Simple: enumerate both robot paths; Better: memoized 3D DP; Best: bottom-up rolling 2D DP

### Time and space

O(rows × cols² × 9) time, O(cols²) best space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def cherry_pickup_memo(grid):
    rows, cols = len(grid), len(grid[0])
    memo = {}

    def f(row, c1, c2):
        if not (0 <= c1 < cols and 0 <= c2 < cols):
            return -10**9
        if row == rows:
            return 0
        key = (row, c1, c2)
        if key in memo:
            return memo[key]

        cherries = grid[row][c1]
        if c1 != c2:
            cherries += grid[row][c2]

        best_next = max(
            f(row+1, c1+d1, c2+d2)
            for d1 in (-1,0,1)
            for d2 in (-1,0,1)
        )
        memo[key] = cherries + best_next
        return memo[key]

    return f(0, 0, cols-1)

grid = [[3,1,1],[2,5,1],[1,5,5],[2,1,1]]
print(cherry_pickup_memo(grid))
```

## 1505. Uncrossed Lines

**Topic:** Algorithms  
**Difficulty:** Medium

Maximum non-crossing matches between two arrays is the Longest Common Subsequence problem.

### Simple → Better → Best

Simple: recursion; Better: 2D DP; Best: 1D rolling DP

### Time and space

Best: O(mn) time, O(n) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def uncrossed_recursive(a, b):
    def f(i, j):
        if i == len(a) or j == len(b):
            return 0
        if a[i] == b[j]:
            return 1 + f(i+1, j+1)
        return max(f(i+1, j), f(i, j+1))
    return f(0, 0)

def uncrossed_dp(a, b):
    dp = [[0]*(len(b)+1) for _ in range(len(a)+1)]
    for i in range(len(a)-1, -1, -1):
        for j in range(len(b)-1, -1, -1):
            if a[i] == b[j]:
                dp[i][j] = 1 + dp[i+1][j+1]
            else:
                dp[i][j] = max(dp[i+1][j], dp[i][j+1])
    return dp[0][0]

def uncrossed_1d(a, b):
    dp = [0]*(len(b)+1)
    for x in a:
        previous = 0
        for j, y in enumerate(b, start=1):
            saved = dp[j]
            if x == y:
                dp[j] = previous + 1
            else:
                dp[j] = max(dp[j], dp[j-1])
            previous = saved
    return dp[-1]

print(uncrossed_1d([1,4,2], [1,2,4]))
```

## 1508. Reconstruct Itinerary

**Topic:** Data Structures  
**Difficulty:** Hard

Use every ticket once and return the lexicographically smallest valid itinerary.

**Pattern:** Eulerian path using Hierholzer's algorithm.

### Simple → Better → Best

Simple: backtracking; Better: backtracking with sorted adjacency; Best: Hierholzer Eulerian path

### Time and space

O(E log E) time, O(E) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def itinerary_backtracking(tickets):
    tickets = sorted(tickets)
    used = [False]*len(tickets)
    route = ["JFK"]

    def dfs():
        if len(route) == len(tickets)+1:
            return True
        current = route[-1]
        for i, (src, dst) in enumerate(tickets):
            if not used[i] and src == current:
                used[i] = True
                route.append(dst)
                if dfs(): return True
                route.pop()
                used[i] = False
        return False

    dfs()
    return route

def itinerary_hierholzer(tickets):
    graph = defaultdict(list)
    for src, dst in sorted(tickets, reverse=True):
        graph[src].append(dst)

    route = []
    def visit(airport):
        while graph[airport]:
            visit(graph[airport].pop())
        route.append(airport)

    visit("JFK")
    return route[::-1]

tickets = [["MUC","LHR"],["JFK","MUC"],["SFO","SJC"],["LHR","SFO"]]
print(itinerary_hierholzer(tickets))
```

## 1528. Determine if String Halves Are Alike

**Topic:** Algorithms  
**Difficulty:** Easy

Split an even-length string into two halves and compare vowel counts.

### Simple → Better → Best

Simple: count each half separately; Better: one-pass balance; Best: same one-pass with constant vowel lookup

### Time and space

O(n) time, O(1) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

VOWELS = set("aeiouAEIOU")

def halves_alike_simple(s):
    mid = len(s)//2
    return sum(ch in VOWELS for ch in s[:mid]) == sum(ch in VOWELS for ch in s[mid:])

def halves_alike_one_pass(s):
    balance = 0
    mid = len(s)//2
    for i, ch in enumerate(s):
        if ch in VOWELS:
            balance += 1 if i < mid else -1
    return balance == 0

print(halves_alike_one_pass("book"))
```

## 1532. Count of Smaller Numbers After Self

**Topic:** Algorithms  
**Difficulty:** Hard

For each position, count later elements that are smaller.

**Pattern:** merge sort counting or Fenwick tree.

### Simple → Better → Best

Simple: nested loops; Better: Fenwick tree; Best: modified merge sort

### Time and space

O(n log n) time, O(n) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def count_smaller_bruteforce(nums):
    return [sum(y < x for y in nums[i+1:]) for i, x in enumerate(nums)]

def count_smaller_merge(nums):
    result = [0]*len(nums)
    indexed = list(enumerate(nums))

    def sort(items):
        if len(items) <= 1:
            return items
        mid = len(items)//2
        left = sort(items[:mid])
        right = sort(items[mid:])
        merged = []
        i = j = 0
        while i < len(left) and j < len(right):
            if left[i][1] <= right[j][1]:
                result[left[i][0]] += j
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[j])
                j += 1
        while i < len(left):
            result[left[i][0]] += j
            merged.append(left[i])
            i += 1
        merged.extend(right[j:])
        return merged

    sort(indexed)
    return result

print(count_smaller_merge([5,2,6,1]))
```

## 1539. Path with Maximum Probability

**Topic:** Data Structures  
**Difficulty:** Medium

Find the path maximizing the product of edge probabilities.

**Pattern:** Dijkstra, but maximize probability instead of minimizing distance.

### Simple → Better → Best

Simple: enumerate paths; Better: Bellman-Ford-style relaxation; Best: max-heap Dijkstra

### Time and space

O((V+E) log V) time, O(V+E) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def max_probability(n, edges, probabilities, start, end):
    graph = defaultdict(list)
    for (u, v), p in zip(edges, probabilities):
        graph[u].append((v, p))
        graph[v].append((u, p))

    best = [0.0]*n
    best[start] = 1.0
    heap = [(-1.0, start)]

    while heap:
        neg_prob, node = heapq.heappop(heap)
        prob = -neg_prob
        if node == end:
            return prob
        if prob < best[node]:
            continue
        for neighbor, edge_prob in graph[node]:
            candidate = prob * edge_prob
            if candidate > best[neighbor]:
                best[neighbor] = candidate
                heapq.heappush(heap, (-candidate, neighbor))
    return 0.0

print(max_probability(
    3,
    [[0,1],[1,2],[0,2]],
    [0.5,0.5,0.2],
    0,
    2
))
```

## 1541. Two Sum

**Topic:** Data Structures  
**Difficulty:** Easy

Find two indices whose values sum to a target.

**How to think:** as you scan, ask whether the complement has already appeared.

### Simple → Better → Best

Simple: nested loops; Better: sort + two pointers; Best: one-pass hash map

### Time and space

Best: O(n) average time, O(n) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def two_sum_bruteforce(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

def two_sum_sort(nums, target):
    indexed = sorted((value, index) for index, value in enumerate(nums))
    left, right = 0, len(indexed) - 1
    while left < right:
        total = indexed[left][0] + indexed[right][0]
        if total == target:
            return [indexed[left][1], indexed[right][1]]
        if total < target:
            left += 1
        else:
            right -= 1
    return []

def two_sum_hash(nums, target):
    seen = {}
    for i, value in enumerate(nums):
        need = target - value
        if need in seen:
            return [seen[need], i]
        seen[value] = i
    return []

nums = [2, 7, 11, 15]
print(two_sum_hash(nums, 9))
```

## 1546. Successful Pairs of Spells and Potions

**Topic:** Algorithms  
**Difficulty:** Medium

For each spell, count potions such that `spell × potion >= success`.

**Pattern:** sort once and binary search the first valid potion.

### Simple → Better → Best

Simple: all pairs; Better: sort + manual binary search; Best: sort + lower_bound

### Time and space

O(m log m + n log m) time, O(m) sort space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def successful_pairs_bruteforce(spells, potions, success):
    return [sum(spell * potion >= success for potion in potions) for spell in spells]

def successful_pairs_binary(spells, potions, success):
    potions = sorted(potions)
    n = len(potions)
    result = []
    for spell in spells:
        need = (success + spell - 1)//spell
        index = bisect.bisect_left(potions, need)
        result.append(n - index)
    return result

print(successful_pairs_binary([5,1,3], [1,2,3,4,5], 7))
```

## 1547. Extra Characters in a String

**Topic:** Data Structures  
**Difficulty:** Medium

Split a string using dictionary words while minimizing unused characters.

**Pattern:** dynamic programming over string positions.

### Simple → Better → Best

Simple: recursion; Better: memoization; Best: bottom-up DP / trie optimization

### Time and space

O(n²) time in standard DP, O(n) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def min_extra_recursive(s, dictionary):
    words = set(dictionary)
    def f(i):
        if i == len(s):
            return 0
        best = 1 + f(i+1)
        for j in range(i+1, len(s)+1):
            if s[i:j] in words:
                best = min(best, f(j))
        return best
    return f(0)

def min_extra_memo(s, dictionary):
    words = set(dictionary)
    memo = {}
    def f(i):
        if i == len(s): return 0
        if i in memo: return memo[i]
        best = 1 + f(i+1)
        for j in range(i+1, len(s)+1):
            if s[i:j] in words:
                best = min(best, f(j))
        memo[i] = best
        return best
    return f(0)

def min_extra_dp(s, dictionary):
    words = set(dictionary)
    dp = [0]*(len(s)+1)
    for i in range(len(s)-1, -1, -1):
        dp[i] = 1 + dp[i+1]
        for j in range(i+1, len(s)+1):
            if s[i:j] in words:
                dp[i] = min(dp[i], dp[j])
    return dp[0]

print(min_extra_dp("leetscode", ["leet","code","leetcode"]))
```

## 1554. Frequency of the Most Frequent Element

**Topic:** Algorithms  
**Difficulty:** Medium

After at most `k` increments, maximize the frequency of one value.

**Pattern:** sort + sliding window using the cost to raise all values to the rightmost value.

### Simple → Better → Best

Simple: try every target; Better: prefix sums; Best: sorted sliding window

### Time and space

O(n log n) time, O(1) extra after sorting

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def max_frequency_bruteforce(nums, k):
    best = 1
    for target in nums:
        costs = sorted(max(0, target - x) for x in nums)
        used = count = 0
        for cost in costs:
            if used + cost > k:
                break
            used += cost
            count += 1
        best = max(best, count)
    return best

def max_frequency_best(nums, k):
    nums.sort()
    left = total = best = 0
    for right, value in enumerate(nums):
        total += value
        while value * (right-left+1) - total > k:
            total -= nums[left]
            left += 1
        best = max(best, right-left+1)
    return best

print(max_frequency_best([1,2,4], 5))
```

## 1568. Next Greater Element I

**Topic:** Data Structures  
**Difficulty:** Easy

For each value in one array, find the first greater value to its right in another array.

**Pattern:** monotonic decreasing stack.

### Simple → Better → Best

Simple: scan right for each value; Better: preprocessing; Best: monotonic stack

### Time and space

Best: O(n+m) time, O(n) space

### What to say in the interview

> “I’ll first clarify the inputs, outputs, and edge cases.  
> The simple solution is ____, costing ____ time and ____ space.  
> Its repeated work is ____. I can improve that with ____.  
> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def next_greater_bruteforce(nums1, nums2):
    out = []
    for x in nums1:
        start = nums2.index(x)
        answer = -1
        for y in nums2[start+1:]:
            if y > x:
                answer = y
                break
        out.append(answer)
    return out

def next_greater_stack(nums1, nums2):
    stack = []
    greater = {}
    for x in nums2:
        while stack and stack[-1] < x:
            greater[stack.pop()] = x
        stack.append(x)
    return [greater.get(x, -1) for x in nums1]

print(next_greater_stack([4,1,2], [1,3,4,2]))
```

## 1569. Shuffle an Array

**Topic:** Math

Produce a uniformly random permutation.

**Pattern:** Fisher–Yates shuffle.

### Simple → Better → Best

### Time and space

Best: O(n) time, O(n) output copy or O(1) in-place

### What to say in the interview

> The key invariant/state is ____.  
> I’ll walk through a normal case and the most important edge case.”


```python

def shuffle_sort_key(nums):
    return [x for _, x in sorted((random.random(), x) for x in nums)]

def shuffle_remove(nums):
    pool = nums[:]
    result = []
    while pool:
        index = random.randrange(len(pool))
        result.append(pool.pop(index))
    return result

def fisher_yates(nums):
    result = nums[:]
    for i in range(len(result)-1, 0, -1):
        j = random.randint(0, i)
        result[i], result[j] = result[j], result[i]
    return result

print(fisher_yates([1,2,3,4,5]))
```
